```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re

import scipy.cluster.hierarchy as shc

%config InlineBackend.figure_format = "retina"
%matplotlib inline
```

```python
data = pd.read_excel("../data/ssc_case_study_2_inflammatory_bowel_disease.xlsx")

# Cleaning descriptive data (includes group, sex, age, ethnicity)
descriptive_data = data.loc[:3]
descriptive_data.set_index("Other Variables", inplace=True)
del descriptive_data["Probe.Set.ID"]
patient_id = [re.search("SAK_(\d)*", s).group(0) for s in descriptive_data.columns]
descriptive_data = descriptive_data.T
descriptive_data.index = patient_id

# Gene dictionary
matched_genes = pd.read_excel("../data/ssc_case_study_2_inflammatory_bowel_disease.xlsx", sheet_name=1)
matched_genes.set_index("Probe.Set.ID", inplace=True)

# Gene expression data!
data = data.loc[4:]
data.set_index("Probe.Set.ID", inplace=True)
del data["Other Variables"]
data = data.T
data.index = patient_id
data.columns = [matched_genes.loc[probe_set]["Gene.Symbol"] for probe_set in data.columns]

descriptive_data.columns.name = None
data.columns.name = None
```

```python
plt.figure(figsize=(10, 7))
a = shc.dendrogram(shc.linkage(data, method='ward'))
```
